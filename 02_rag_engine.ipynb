```python
from qdrant_client import QdrantClient
from qdrant_client.http import models
from rag_engine.core.config import get_settings

settings = get_settings()

# Khởi tạo client
client = QdrantClient(url=settings.QDRANT_URL, api_key=settings.QDRANT_API_KEY)

# Danh sách tất cả các trường metadata bạn vừa tìm thấy (trừ 'content' vì nó quá dài, không dùng để filter)
fields_to_index = [
    'source_file', 'document_name', 'so_hieu', 'loai_van_ban', 
    'ngay_ban_hanh', 'chuong', 'ten_chuong', 'muc', 'ten_muc', 
    'dieu', 'ten_dieu', 'khoan', 'diem', 'cap_chunk', 'chunk_id'
]

print("⏳ Đang thiết lập Index cho Qdrant Cloud...")

for field in fields_to_index:
    try:
        client.create_payload_index(
            collection_name=settings.QDRANT_COLLECTION,
            field_name=field,
            field_schema=models.PayloadSchemaType.KEYWORD,
        )
        print(f"✅ Đã xong: {field}")
    except Exception:
        # Nếu đã có index rồi thì bỏ qua
        print(f"ℹ️ Bỏ qua: {field} (đã tồn tại hoặc không cần thiết)")

print("\n🔥 TẤT CẢ ĐÃ SẴN SÀNG! Hệ thống RAG của bạn giờ đã 'thông kinh mạch'.")
```

```python
import nest_asyncio
import uvicorn
import requests
import time

nest_asyncio.apply()

# Lấy URL Ngrok (Thử lại 3 lần)
public_url = "Chưa xác định"
for _ in range(3):
    try:
        # Port 4040 là port mặc định của ngrok dashboard
        res = requests.get("http://127.0.0.1:4040/api/tunnels", timeout=2)
        public_url = res.json()['tunnels'][0]['public_url']
        break
    except:
        time.sleep(1)

from rag_engine.api.main import app

print(f"✅ Ngrok URL: {public_url}")
print(f"🚀 Swagger UI: {public_url}/docs")

config = uvicorn.Config(
    app, 
    host="0.0.0.0", 
    port=8001, 
    log_level="info"
)
server = uvicorn.Server(config)
await server.serve()
```
